# Support Vector Machine for Classification

SVM is a supervised machine learning algorithm which can be used for both
classification or regression challenges. However, it is mostly used in
classification problems. In the SVM algorithm, we plot each data item as a point
in n-dimensional space (where n is number of features you have) with the value of
each feature being the value of a particular coordinate. Then, we perform
classification by finding the hyper-plane that differentiate the two classes
very well.

[Geeks for Geeks](https://www.geeksforgeeks.org/image-classification-using-support-vector-machine-svm-in-python/)

```python
# Standard Library
import os

# Dependencies
# Data Manipulation
import pandas as pd
from cv2 import imshow
# Image Manipulation
import skimage as skimg
## Plotting
import matplotlib.pyplot as plt
# Math
import numpy as np
# Machine Learning
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# Custom
from data.properties import read_properties
```

```python
categories = ['pokemon','digimon']
img_input = [] #input array
img_output = [] #output array
data_dir = read_properties('paths.env')['dataset-dir']
source_images = os.path.join(data_dir, 'images')
```

```python
limit = 152
for category in categories:
    print(f'loading... category : {category}')
    category_path = os.path.join(source_images, category)
    for img_name in os.listdir(category_path):
        print(f"Loading: {img_name}")
        if img_input.__len__() == limit:
            limit += limit
            break
        img_path = os.path.join(category_path, img_name)
        img = skimg.io.imread(img_path)
        img = skimg.transform.resize(img, (100, 100))
        img_input.append(img.flatten())
        img_output.append(categories.index(category))
    print(f'loaded category:{category} successfully, input size: {img_input.__len__()}')
```
